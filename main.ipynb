```python
import matplotlib.pyplot as plt
```

```python
#function to get the minimum  value between 3 values
def minimum (x, y, z):
    if (x < y):
        return x if (x < z) else z
    else:
        return y if (y < z) else z
```

```python
#showing the matrix as a 3x3 image graph with values in each index
def plottingGraph(matrix):

	plt.matshow(matrix)
	plt.annotate(matrix[0][0],(0,0), size="17", color="white", ha="center", va="center")
	plt.annotate(matrix[0][1],(1,0),size="17", color="white", ha="center", va="center")
	plt.annotate(matrix[0][2],(2,0),size="17", color="white", ha="center", va="center")
	plt.annotate(matrix[1][0],(0,1),size="17", color="white", ha="center", va="center")
	plt.annotate(matrix[2][0],(0,2),size="17", color="white", ha="center", va="center")
	plt.annotate(matrix[1][1],(1,1),size="17", color="white", ha="center", va="center")
	plt.annotate(matrix[1][2],(2,1),size="17", color="white", ha="center", va="center")
	plt.annotate(matrix[2][1],(1,2),size="17", color="white", ha="center", va="center")
	plt.annotate(matrix[2][2],(2,2),size="17", color="white", ha="center", va="center")
```

```python
def TopRightBottomLeft(matrix):
    firstPathCost = matrix[0][0] + matrix[0][1] + matrix[1][2] + matrix[2][2]
    secondPathCost = matrix[0][0] + matrix[1][1] + matrix[2][2]
    thirdPathCost = matrix[0][0] + matrix[1][0] + matrix[2][1] + matrix[2][2]

    #First Path Graph 
    if minimum(firstPathCost, secondPathCost, thirdPathCost) == firstPathCost:
      plt.plot((0,1),(0,0),marker="o")
      plt.plot((1,2),(0,1), marker="o")
      plt.plot((1,2),(0,1), marker="o")
      plt.plot((2,2),(1,2), marker="o")

    #Second Path Plot  
    elif minimum(firstPathCost, secondPathCost, thirdPathCost) == secondPathCost:
      plt.plot((1,0),(1,0),marker="o")
      plt.plot((1,2),(1,2), marker="o")
   
    #Third Path Graph:
    elif minimum(firstPathCost, secondPathCost, thirdPathCost) == thirdPathCost:
      #ThirdPathGraph
      plt.plot((0,0),(0,1),marker="o")
      plt.plot((0,1),(1,2), marker="o")
      plt.plot((1,2),(2,2), marker="o")
```

```python
def BottomLeftTopRight(matrix):
    firstPathCost = matrix[2][0] + matrix[1][0] + matrix[0][1] + matrix[0][2]
    secondPathCost = matrix[2][0] + matrix[1][1] + matrix[0][2]
    thirdPathCost = matrix[2][0] + matrix[2][1] + matrix[1][2] + matrix[0][2]

    #First Path Graph 
    if minimum(firstPathCost, secondPathCost, thirdPathCost) == firstPathCost:
      plt.plot((0,0),(2,1),marker="o")
      plt.plot((0,1),(1,0), marker="o")
      plt.plot((1,2),(0,0), marker="o")
  
    #Second Path Plot  
    elif minimum(firstPathCost, secondPathCost, thirdPathCost) == secondPathCost:
      plt.plot((0,1),(2,1),marker="o")
      plt.plot((1,2),(1,0), marker="o")
   
    #Third Path Graph:
    elif minimum(firstPathCost, secondPathCost, thirdPathCost) == thirdPathCost:
    
      plt.plot((0,1),(2,2),marker="o")
      plt.plot((1,2),(2,1), marker="o")
      plt.plot((2,2),(1,0), marker="o")
```

```python
#Test Cases:
matrix1 = [[8, 24, 3],
           [67, 4, 2],
           [2, 0, 7]]

matrix2 = [[9,3,70],
           [83,120,15],
           [5,2,33]]

matrix3 = [[1,2,8],
           [3,6,11],
           [10,20,16]]
```

```python
#minimuim cost path for matrix 1
plottingGraph(matrix1)
TopRightBottomLeft(matrix1)
plottingGraph(matrix1)
BottomLeftTopRight(matrix1)
```

```python
#minimuim cost path for matrix 2
plottingGraph(matrix2)
TopRightBottomLeft(matrix2)
plottingGraph(matrix2)
BottomLeftTopRight(matrix2)
```

```python
#minimum cost path for matrix 3
plottingGraph(matrix3)
TopRightBottomLeft(matrix3)
plottingGraph(matrix3)
BottomLeftTopRight(matrix3)
```
